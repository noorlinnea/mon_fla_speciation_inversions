```python
import demes
import demesdraw
from matplotlib import pyplot as plt
```

```python
model_graph = demes.load("/Users/noorapoikela/Desktop/demesdraw/Dmon_Dfla_IM.yaml")
```

```python
font = {'family' : 'DejaVu Sans', 'weight' : 'normal', 'size' : 15}
plt.rc('font', **font)
```

```python
demesdraw.tubes(model_graph, max_time=4e6, colours={"ancestor":"#99CCCC", "Dmontana":"#6666FF", "Dflavomontana":"#FF9900"}, positions = {"ancestor":1.5e6, "Dmontana":1e6, "Dflavomontana":2e6})
plt.savefig("/Users/noorapoikela/Desktop/demesdraw/Dmon_Dfla_IM.png", facecolor='white', dpi=300)
plt.show()
```
